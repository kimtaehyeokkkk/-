```python
#2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def f(x,y):
	return x**2 + y**2 +4* np.sin(x+y)

def grad_f(x,y):
  h=3
  return [(f(x+h,y) - f(x-h,y))/(2*h),(f(x,y+h) - f(x,y-h))/(2*h)]


def gradient_descent(theta0, iters, alpha):
	history = [theta0] 
	theta = theta0    
	for i in range(iters):
		gradient = grad_f(theta[0],theta[1])
		theta = [x - alpha*y for x,y in zip(theta, gradient)]
		history.append(theta)
	return history

history = gradient_descent(theta0 = [-3, -3], iters =1, alpha = 0.01)

fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

x = np.arange(-3, 3, 0.5)
y = np.arange(-3, 3, 0.5)
x, y = np.meshgrid(x, y)
z = f(x,y)
surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, alpha=0.25, 
		 linewidth=0, antialiased=False, cmap='rainbow')


x = np.array([a[0] for a in history])
y = np.array([a[1] for a in history])
z = f(x,y)
ax.scatter(x, y, z, color="r"); 

print(z)


plt.xlabel('Feature A')
plt.ylabel('Feature B')


plt.show()
```
